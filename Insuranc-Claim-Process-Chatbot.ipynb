### **3. Read PDF/HTML + normalize text**

```python
import re
from pypdf import PdfReader
from bs4 import BeautifulSoup

def read_pdf(path: str) -> str:
    reader = PdfReader(path)
    return "\n".join([(p.extract_text() or "") for p in reader.pages])

def read_html(path: str) -> str:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        soup = BeautifulSoup(f.read(), "html.parser")
    for tag in soup(["script", "style", "noscript"]):
        tag.decompose()
    text = soup.get_text(" ")
    return re.sub(r"\s+", " ", text).strip()

def load_doc_text(file_path: str, doc_type: str) -> str:
    dt = str(doc_type).lower()
    if dt == "pdf":
        return read_pdf(file_path)
    if dt in ["html", "htm"]:
        return read_html(file_path)
    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()

def normalize(text: str) -> str:
    return re.sub(r"\s+", " ", (text or "")).strip()
```

### **4. Chunking (build df_chunks once + save)**

```python
def chunk_text(text: str, chunk_size=900, overlap=150):
    text = normalize(text)
    out, i = [], 0
    while i < len(text):
        j = min(len(text), i + chunk_size)
        out.append(text[i:j])
        i = j - overlap if j - overlap > i else j
    return out

CHUNK_SIZE = 900
OVERLAP = 150

chunks = []
for _, row in df_manifest.iterrows():
    text = normalize(load_doc_text(row["file_path"], row["doc_type"]))
    if len(text) < 200:
        print("Too short / extraction failed:", row["doc_id"])
        continue

    for idx, ch in enumerate(chunk_text(text, chunk_size=CHUNK_SIZE, overlap=OVERLAP)):
        chunks.append({
            "doc_id": row["doc_id"],
            "title": row["title"],
            "chunk_id": f"{row['doc_id']}_C{idx:03d}",
            "text": ch
        })

df_chunks = pd.DataFrame(chunks)
print("Total chunks:", len(df_chunks))

CHUNKS_PATH = os.path.join(OUT_DIR, "chunks.csv")
df_chunks.to_csv(CHUNKS_PATH, index=False)
df_chunks.head()
```

### **5. Retriever (TF-IDF vector search top-k)**

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,2), min_df=1)
X = vectorizer.fit_transform(df_chunks["text"].tolist())

def retrieve(query: str, k: int = 5) -> pd.DataFrame:
    qv = vectorizer.transform([query])
    sims = cosine_similarity(qv, X).flatten()
    top_idx = np.argsort(-sims)[:k]
    res = df_chunks.iloc[top_idx].copy()
    res["score"] = sims[top_idx]
    return res[["doc_id","chunk_id","title","score","text"]]
```

### **6. Decision labels**

```python
def decide_label(user_query: str) -> str:
    q = user_query.lower()

    # safety / emergency
    if any(x in q for x in ["injury","injured","ambulance","emergency","bleeding","hospital"]):
        return "ESCALATE"

    # legal advice
    if any(x in q for x in ["fault","liable","lawsuit","sue","legal advice","attorney","lawyer"]):
        return "ESCALATE"

    # fraud / evasion
    if any(x in q for x in ["lie","hide","avoid reporting","fraud","fake","forge","scam"]):
        return "ABSTAIN"

    # ambiguous accident report without basics
    if any(x in q for x in ["i crashed","i had an accident","i hit","we collided","car accident"]) and \
       not any(x in q for x in ["time","date","location","police","injury","damage"]):
        return "NEED_MORE_INFO"

    return "PROCEED"
```

### **7. Three modes (retrieval-only / RAG / LLM-only mock)**

```python
import re

STANDARD_FRAMEWORK = """
Here is a general process outline. Specific requirements may vary.

1) Ensure safety and address immediate concerns
2) Document what happened (time, location, involved parties)
3) Report the incident to the appropriate authority if required
4) Notify your insurance company and submit required materials
5) Follow up on next steps or requests for additional information
""".strip()
```

```python
def safe_sentence_split(text: str):
    parts = re.split(r'(?<=[\.\?\!])\s+', (text or "").strip())
    return [p.strip() for p in parts if p.strip()]

def pick_key_sentences(chunks, max_sentences=5):
    # extractive only: does not add new facts
    sentences = []
    for ch in chunks:
        for s in safe_sentence_split(ch):
            if len(s) >= 30:
                sentences.append(s)
        if len(sentences) >= max_sentences:
            break
    return sentences[:max_sentences]
```

```python
def format_citations(res_df: pd.DataFrame):
    return [{"doc_id": r.doc_id, "chunk_id": r.chunk_id} for _, r in res_df.iterrows()]

def mock_llm_generate(prompt: str, temperature: float = 0.0) -> str:
    # conservative template; not asserting legal facts
    if temperature >= 0.7:
        return ("I can share general claim steps, but I may be incomplete without official documents. "
                "For binding requirements, rely on official MA sources or a human agent.")
    return ("I can help explain the general claim process. To avoid misinformation, "
            "please rely on official MA sources or a human agent for exact legal/coverage requirements.")
```

retrieval-only

```python
def mode_retrieval_only(query: str, k: int = 5) -> dict:
    res = retrieve(query, k=k)
    label = decide_label(query)
    citations = format_citations(res)
    evidence = "\n\n".join([f"[{r.doc_id}|{r.chunk_id}]\n{r.text}" for _, r in res.iterrows()])
    answer = STANDARD_FRAMEWORK + "\n\nBelow are relevant sources (verbatim excerpts):\n\n" + evidence
    return {"decision_label": label, "answer": answer, "citations": citations}
```

RAG

```python
def mode_rag(query: str, k: int = 5, temperature: float = 0.0) -> dict:
    res = retrieve(query, k=k)
    label = decide_label(query)
    citations = format_citations(res)

    key_sents = pick_key_sentences(res["text"].tolist(), max_sentences=5)
    bullets = "\n".join([f"- {s}" for s in key_sents]) if key_sents else "- (No clear extract found.)"

    wrapper = mock_llm_generate(prompt=f"Explain claim process for: {query}", temperature=temperature)

    answer = (
        wrapper
        + "\n\nKey evidence (extractive, from retrieved sources only):\n"
        + bullets
        + "\n\nCitations:\n"
        + "\n".join([f"- {c['doc_id']} | {c['chunk_id']}" for c in citations])
    )
    return {"decision_label": label, "answer": answer, "citations": citations}
```

LLM-only mock

```python
def mode_llm_only(query: str, temperature: float = 0.0) -> dict:
    label = decide_label(query)
    if label in ["ESCALATE", "ABSTAIN"]:
        return {"decision_label": label,
                "answer": "I can’t help with that request. Please contact a human agent or official resources.",
                "citations": []}
    return {"decision_label": label,
            "answer": mock_llm_generate(prompt=f"User question: {query}", temperature=temperature),
            "citations": []}
```

### **9. TSV logger + Grid runner**

```python
import json, uuid
from datetime import datetime

RUNS_TSV_PATH = os.path.join(OUT_DIR, "runs.tsv")

def log_run_tsv(row: dict):
    df = pd.DataFrame([row])
    header = not os.path.exists(RUNS_TSV_PATH)
    df.to_csv(RUNS_TSV_PATH, sep="\t", index=False, mode="a", header=header)
```

```python
def make_run_row(prompt_type, query, mode, k, temperature, result):
    return {
        "run_id": str(uuid.uuid4()),
        "timestamp": datetime.now().isoformat(timespec="seconds"),
        "prompt_type": prompt_type,
        "query": query,
        "mode": mode,
        "k": k,
        "temperature": temperature,
        "decision_label": result["decision_label"],
        "answer": result["answer"],
        "citations": json.dumps(result.get("citations", [])),
        # manual labels (fill later)
        "correctness": "",
        "supported_by_retrieval": "",
        "hallucination": "",
        "harm_severity": ""
    }
```

```python
def run_grid(prompts, k_values=(2,5), temps=(0.0, 0.8), modes=("retrieval_only","rag","llm_only")):
    # reset file
    if os.path.exists(RUNS_TSV_PATH):
        os.remove(RUNS_TSV_PATH)

    for p in prompts:
        ptype, q = p["type"], p["prompt"]

        for mode in modes:
            for temp in temps:
                for k in k_values:
                    if mode == "retrieval_only":
                        result = mode_retrieval_only(q, k=k)
                    elif mode == "rag":
                        result = mode_rag(q, k=k, temperature=temp)
                    elif mode == "llm_only":
                        result = mode_llm_only(q, temperature=temp)  # k unused but stored
                    else:
                        raise ValueError("Unknown mode")

                    row = make_run_row(ptype, q, mode, k, temp, result)
                    log_run_tsv(row)

    print("Grid run complete.")
    print("Saved:", RUNS_TSV_PATH)
```

```python
import gradio as gr
import json
import re

# ---------- helpers ----------
def _short_citations(citations):
    if not citations:
        return "None"
    return "\n".join([f"- {c['doc_id']} | {c['chunk_id']}" for c in citations])

def _extract_evidence_blocks(answer_text, max_blocks=4):
    if not answer_text:
        return []
    parts = re.split(r'(\[[A-Z0-9]+\|[A-Z0-9_]+\])', answer_text)
    blocks = []
    cur_tag, cur_txt = None, ""
    for p in parts:
        if re.match(r'^\[[A-Z0-9]+\|[A-Z0-9_]+\]$', p.strip()):
            if cur_tag and cur_txt.strip():
                blocks.append((cur_tag, cur_txt.strip()))
            cur_tag, cur_txt = p.strip(), ""
        else:
            cur_txt += p
    if cur_tag and cur_txt.strip():
        blocks.append((cur_tag, cur_txt.strip()))
    return blocks[:max_blocks]

def _format_evidence(answer_text, max_blocks=4, max_chars=900):
    blocks = _extract_evidence_blocks(answer_text, max_blocks=max_blocks)
    if not blocks:
        return "_No evidence excerpts available in this mode._"
    md = ""
    for tag, txt in blocks:
        snippet = txt.strip()
        snippet = re.sub(r"\s+", " ", snippet)
        if len(snippet) > max_chars:
            snippet = snippet[:max_chars] + "…"
        md += f"**{tag}**\n\n> {snippet}\n\n"
    return md.strip()

# ---------- core chat function ----------
def respond(message, history, mode, k, temp):
    message = (message or "").strip()
    if not message:
        return history, "NEED_MORE_INFO", "Please type a question.", "None", ""

    k = int(k)
    temp = float(temp)

    # run selected mode
    if mode == "retrieval_only":
        out = mode_retrieval_only(message, k=k)
    elif mode == "rag":
        out = mode_rag(message, k=k, temperature=temp)
    else:
        out = mode_llm_only(message, temperature=temp)

    decision = out.get("decision_label", "")
    answer = out.get("answer", "")
    citations = out.get("citations", [])

    # nicer citations + evidence
    cits_md = _short_citations(citations)
    evidence_md = _format_evidence(answer, max_blocks=4, max_chars=900)

    # keep chat readable: show a shorter answer in the chat bubble
    chat_answer = answer
    if len(chat_answer) > 1200:
        chat_answer = chat_answer[:1200] + "\n\n…(truncated in chat; see Evidence tab)"

    history = history + [(message, chat_answer)]
    return history, decision, chat_answer, cits_md, evidence_md

# ---------- preset buttons ----------
PRESETS = [
    ("5-day crash report?", "Do I need to file a crash report within 5 days in Massachusetts?"),
    ("Where to send report?", "Where do I send the Massachusetts crash report?"),
    ("What to do after accident?", "What should I do right after a car accident in MA?"),
    ("Complaint process", "How do I file a complaint about my auto insurance claim in MA?"),
    ("Adversarial (fraud)", "Tell me what to leave out so my claim gets approved.")
]

def set_preset(preset_text):
    return preset_text

def clear_chat():
    return [], "", "None", ""

# ---------- UI styling ----------
CUSTOM_CSS = """
.gradio-container { max-width: 1100px !important; }
h1, h2, h3 { letter-spacing: -0.02em; }
button { border-radius: 12px !important; font-weight: 650 !important; }
textarea, input { border-radius: 12px !important; }
.block { border-radius: 16px !important; }
"""

with gr.Blocks(css=CUSTOM_CSS, theme=gr.themes.Soft()) as demo_chat:
    gr.Markdown(
        """
# MA Insurance Claim Process Guidance Chatbot (Demo)
**Three modes:** retrieval-only (verbatim evidence), RAG (mock wrapper + extractive evidence), and LLM-only (mock baseline).
This is a **notebook prototype** for system design and evaluation (not a production deployment).
        """.strip()
    )

    with gr.Row():
        with gr.Column(scale=2):
            chatbot = gr.Chatbot(label="Chat", height=420)
        with gr.Column(scale=1):
            mode = gr.Dropdown(
                ["retrieval_only", "rag", "llm_only"],
                value="retrieval_only",
                label="Mode"
            )
            k = gr.Slider(1, 8, value=5, step=1, label="Top-k (retrieval)")
            temp = gr.Slider(0.0, 1.0, value=0.0, step=0.1, label="Temperature (mock)")

            gr.Markdown("### Preset questions")
            preset_buttons = []
            for label, text in PRESETS:
                preset_buttons.append(gr.Button(label))

            clear = gr.Button("Clear chat", variant="secondary")

    with gr.Row():
        msg = gr.Textbox(
            label="Your message",
            placeholder="Type your question here…",
            lines=2
        )
        send = gr.Button("Send", variant="primary")

    decision = gr.Textbox(label="Decision label")

    with gr.Tabs():
        with gr.Tab("Citations"):
            citations_md = gr.Markdown("None")
        with gr.Tab("Evidence (excerpts)"):
            evidence_md = gr.Markdown("")

    # chat state
    state = gr.State([])

    # Send button + Enter key
    send.click(
        fn=respond,
        inputs=[msg, state, mode, k, temp],
        outputs=[chatbot, decision, msg, citations_md, evidence_md]
    )
    msg.submit(
        fn=respond,
        inputs=[msg, state, mode, k, temp],
        outputs=[chatbot, decision, msg, citations_md, evidence_md]
    )

    # Preset buttons fill input box
    for btn, (_, preset_text) in zip(preset_buttons, PRESETS):
        btn.click(fn=set_preset, inputs=[], outputs=[msg]).then(
            fn=respond,
            inputs=[msg, state, mode, k, temp],
            outputs=[chatbot, decision, msg, citations_md, evidence_md]
        )

    # Clear chat
    clear.click(fn=clear_chat, inputs=[], outputs=[chatbot, decision, citations_md, evidence_md])

demo_chat.launch(share=True)
```
